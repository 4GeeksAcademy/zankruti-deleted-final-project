```python
import pandas as pd
import os

# Directory where the CSV files are stored
directory = './'

# List to hold all the DataFrames
dataframes = []

# Assuming the first file starts from January 2019 and you want to read through December 2024
# Update these dates based on your actual file range
start_year = 2016
end_year = 2023

# Loop through each year and month
for year in range(start_year, end_year + 1):
    for month in range(1, 13):
        filename = f'uda_contractor_{year}{month:02}.csv'  # Formats month as two digits
        file_path = os.path.join(directory, filename)

        if os.path.exists(file_path):  # Check if the file exists before reading
            print(f'loading file {filename}')
            
            try:
                df = pd.read_csv(file_path)
            except Exception:
                df = pd.read_csv(file_path, encoding = "ISO-8859-1")
            dataframes.append(df)
        else:
            print(f"File not found: {filename}")

# Concatenate all the DataFrames vertically
combined_df = pd.concat(dataframes, axis=0, ignore_index=True)
```
